## 3.Boxplot

```python
# A. motivating example
y1=[75,75,76,76,77,77,78,79,79,98] # A선생님에게 통계학을 배운 학생의 점수들
y2=[76,76,77,77,78,78,79,80,80,81] # B선생님에게 통계학을 배운 학생의 점수들
```

```python
np.mean(y1)
```

```python
np.mean(y2)
```

```python
# B. matplotlib으로 boxplot 그리기
plt.boxplot(y1);
```

```python
plt.boxplot(y2);
```

```python
# A반 학생들의 점수와 B반 학생들의 점수를 나란히 박스플랏으로 그리자.
plt.boxplot([y1, y2]);
```
